### Pandas data types

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'Lists': [[1, 2, 3], [4, 5, 6], [7, 8, 9]],
    'Sets': [{1, 2}, {3, 4}, {5, 6}],
    'Integers': [10, 20, 30],
    'Floats': [1.5, 2.5, 3.5],
    'Strings': ['a', 'b', 'c'],
    'Booleans': [True, False, True]
})
df
```

```python
df.info()
```

```python
df.dtypes
```

```python
print(df['Lists'].dtype)
```

```python
df = pd.DataFrame(
    {
        'Integers': [10, 20, 30],
        'Floats': [1.5, 2.5, 3.5],
        'Booleans': [True, False, True]
    },
    dtype=np.float32
)
df
```

### Getting columns based on data type

```python
df = pd.DataFrame({
    'Lists': [[1, 2, 3], [4, 5, 6], [7, 8, 9]],
    'Sets': [{1, 2}, {3, 4}, {5, 6}],
    'Integers': [10, 20, 30],
    'Floats': [1.5, 2.5, 3.5],
    'Strings': ['a', 'b', 'c'],
    'Booleans': [True, False, True]
})
df
```

```python
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
numerical_columns
```
